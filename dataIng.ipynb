# Data ingestion from the CMU Movie Summary Corpus.
This dataset integrates Wikipedia plot summaries with metadata 

from Freebase (e.g., character gender, actor ethnicity, movie genre).

Source: http://www.cs.cmu.edu/~ark/personas/

```python
import os
import tarfile
import urllib.request
import pandas as pd

# Definizione percorsi
DATA_DIR = '../data'
URL = "http://www.cs.cmu.edu/~ark/personas/data/MovieSummaries.tar.gz"
ARCHIVE_PATH = os.path.join(DATA_DIR, "MovieSummaries.tar.gz")

# Crea la cartella dati se non esiste
if not os.path.exists(DATA_DIR):
    os.makedirs(DATA_DIR)

# 1. Download del dataset
if not os.path.exists(ARCHIVE_PATH):
    print("Downloading dataset...")
    urllib.request.urlretrieve(URL, ARCHIVE_PATH)
    print("Download complete.")

# 2. Estrazione del file
if not os.path.exists(os.path.join(DATA_DIR, "MovieSummaries")):
    print("Extracting files...")
    with tarfile.open(ARCHIVE_PATH, "r:gz") as tar:
        tar.extractall(path=DATA_DIR)
    print("Extraction complete.")

# 3. Caricamento dei dati (Plot Summaries)
# Il file è un .txt con ID film e trama separati da tab
plots_path = os.path.join(DATA_DIR, 'MovieSummaries', 'plot_summaries.txt')
plots = pd.read_csv(plots_path, sep='\t', names=['wiki_id', 'plot'])

print(f"Dataset caricato: {len(plots)} trame trovate.")
print(plots.head())
```
